```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
from skimage import io as io_url
```

```python
image = io_url.imread('https://github.com/Mian56/CSC467_Week2_Lab/blob/main/Lena.png?raw=true')
frame = np.float32(image) / 255

plt.imshow(frame)
```

```python
print(frame.shape)
```

```python

frm = frame[0:100,0:100,0]
plt.imshow(frm, cmap = "gray")
plt.show()

```

```python
array = np.array(frm)
```

```python
brightened_array = np.clip(array + 0.2, 0, 1)
```

```python
plt.imshow(brightened_array, cmap='gray')
plt.show()
```

```python
mean_kernel = np.ones((3,3))/9
```

```python
#convolution

mean_filtered = cv2.filter2D(frame, -1, mean_kernel)
```

```python
plt.imshow(mean_filtered)
plt.show()
```

```python
sharpen_kernel = np.array([[0,-1,0],[-1,5,-1], [0,-1,0]])

sharpened_image = cv2.filter2D(frame, -1, sharpen_kernel)

plt.imshow(sharpened_image)
plt.show()
```

```python
gray_image = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)

sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize= 3)

sobel_y= cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize= 3)

edges = cv2.magnitude(sobel_x, sobel_y)

edges_normalized = np.clip(edges / np.max(edges),0,1)

plt.imshow(edges_normalized, cmap = 'gray')
plt.show()
```
